## Loading the Updated Inventory JSON File

```python
fd = open('record.json','r')
r = fd.read()
fd.close()
```

```python
import json
```

```python
record = json.loads(r)
```

```python
#to start the sales file length from 0 for using len(sales)+1
sales={}                       
```

## Purchasing Products and Bill generation

```python
#taking input from the user to get details about what he/she wants to buy
num = int(input('Enter number of items you want to buy: '))
print('***Choose the product ids corresponding to the items u want to buy***')
print('\n')
for i in record.keys():
   print('{}-{}'.format(i,record[i]['product_name']))

total_price = 0

for i in range(num):
    
  prod_id=str(input('Enter product id: '))
  p_quantity=int(input('Enter product quantity: '))
  

  if int(prod_id) >=1001 and int(prod_id) <=1030 and p_quantity <= record[prod_id]['quantity']:
    print('\n')
    print('************************')
    
    #to get the time of buy
    import time
    time=time.ctime()
    print(time)
    
    print('************************')
 
    print('Name of the product: ',record[prod_id]['product_name'])
    print('Quantity available: ',record[prod_id]['quantity'])
    print('Price per item: ',record[prod_id]['unit_price'])
    print('Discount: ',record[prod_id]['discount'])
    
    print('************************')
    
    pad = record[prod_id]['unit_price'] - record[prod_id]['discount']
    print('Price after discount: ',pad)
    
    price = pad*p_quantity
    total_price += price
    print('Price for the item: ',price)
  
    print('************************')
    print('************************')
    print('\n')
    
    
    #getting the new quantity after each buy
    record[prod_id]['quantity'] = record[prod_id]['quantity'] - p_quantity
    
    
    #calculate and updating sales details
    import json
    a=len(sales)+1
    sales[a] = {'prod_name': record[prod_id]['product_name'], 'price_after_disc': pad, 'quantity': p_quantity, 'purchase time': time}



  else:
    print('\n')
    print('*******************************************')
    print('              *****SORRY*****')
    print('The product you want to buy is not available')
    print('                     or')
    print('The required quantity of that Product is unavailable')
    print('*******************************************')
    print('\n')
    

print('************************')
print('************************')
print('\n')
print('Total price for all items: ',total_price)
print('\n')
print('************************')
print('************************')


#laoding the updated quantity data into record.json file

import json
js=json.dumps(record)
f=open('record.json','w')
f.write(js)
f.close()
```

## Updating sales.json file

```python
import json

x= open('sales.json','w')
ss= json.dumps(sales)
x.write(ss)
x.close()
```
